## Datos

### Comprobacion que el cálculo de la recuperación sea correcto 

Voy a comprobar si la medición de la recuperación del concentrado final (`rougher.output.recovery`) coincide con la medición del concentrado de oro en la alimentación antes de la flotación (`rougher.input.feed_au`) y  la proporción de oro en las colas rougher justo después de la flotación(`rougher.output.tail_au`).

Para ello usare la formula:   Recuperación = (C * (F - T)) / (F * (C - T)) * 100

```python
# Calcular la recuperación de la característica rougher.output.recovery
C = df_train['rougher.output.concentrate_au']
F = df_train['rougher.input.feed_au']
T = df_train['rougher.output.tail_au']

recovery = (C * (F - T)) / (F * (C - T)) * 100

# Comparar los resultados con los valores reales
recovery_actual = df_train['rougher.output.recovery']

# Calcular el error absoluto medio (EAM)
eam = abs(recovery - recovery_actual).mean()

# Mostrar los resultados
print("Error absoluto medio (EAM):", eam)
```

El error es casi nulo, por lo que puedo estar seguro que la medición de la concentración de oro final es precisa.

Hay que tener en cuenta que tanto los calculos de la recuperacion total como la del calculo esta echa en un dataframe con valores nulos, si los relleno el valor seria diferente ya que `rougher.output.recovery` esta calculado con los valores nulos.

### Calculos de sMAPE y validación cruzada de los modelos

Voy a realizar la medición de `sMAPE` para luego poder calcular el `sMAPE Ponderado` para cada uno de mis modelos, para así poder comparar cual de ellos es el mejor a la hora de evaluar la obtención del concentrado final del oro.

Creare 3 modelos de predición:
- Regreción Lineal
- Arbol de decisiones
- Bosque Aleatorio

En cada uno de los modelos realizare una validación cruzada, para asegurarme de elegir el mejor modelo a la hora de calcular el sMAPE.

```python
def calcular_sMAPE(y_true, y_pred):
    smape_individual = 200 * np.abs(y_true - y_pred) / (np.abs(y_true) + np.abs(y_pred))
    smape_individual = smape_individual[~np.isnan(smape_individual)]
    smape_final = np.mean(smape_individual)
    return smape_final

# Definir las columnas de destino
columnas_objetivo = ['rougher.output.concentrate_au', 'final.output.concentrate_au']

# Obtener las características y la variable objetivo para el conjunto de entrenamiento
X_train = df_train.drop(columns=columnas_objetivo)
y_train_rougher = df_train['rougher.output.concentrate_au']
y_train_final = df_train['final.output.concentrate_au']

# Obtener las características y la variable objetivo para el conjunto de prueba
X_test = df_test.drop(columns=columnas_objetivo)
y_test_rougher = df_test['rougher.output.concentrate_au']
y_test_final = df_test['final.output.concentrate_au']

# Definir los modelos a evaluar
modelos = [
    LinearRegression(),
    DecisionTreeRegressor(),
    RandomForestRegressor()
]

for modelo in modelos:
    # Calcular sMAPE para rougher.output.recovery
    scores_rougher = cross_val_score(modelo, X_train, y_train_rougher, scoring=make_scorer(calcular_sMAPE), cv=5)
    mejor_modelo_rougher = modelo.fit(X_train, y_train_rougher)
    y_pred_rougher = mejor_modelo_rougher.predict(X_test)
    smape_rougher = calcular_sMAPE(y_test_rougher, y_pred_rougher)

    # Calcular sMAPE para final.output.recovery
    scores_final = cross_val_score(modelo, X_train, y_train_final, scoring=make_scorer(calcular_sMAPE), cv=5)
    mejor_modelo_final = modelo.fit(X_train, y_train_final)
    y_pred_final = mejor_modelo_final.predict(X_test)
    smape_final = calcular_sMAPE(y_test_final, y_pred_final)

    # Calcular sMAPE final ponderado
    smape_final_ponderado = 0.25 * smape_rougher + 0.75 * smape_final

    print(type(modelo).__name__)
    print("sMAPE rougher:", smape_rougher)
    print("sMAPE final:", smape_final)
    print("sMAPE final ponderado:", smape_final_ponderado)
    print("Cross-validation scores (rougher):", scores_rougher)
    print("Cross-validation scores (final):", scores_final)
    print("Mean sMAPE (rougher):", np.mean(scores_rougher))
    print("Mean sMAPE (final):", np.mean(scores_final))
    print()
```

El mejor modelo para calcular la obtención del concentrado de oro final es el de Bosque Aleatorio ya que tiene el menor valor de sMAPE con un valor igual a 21.83
